# LSTM

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import LSTM, Dense
from keras.optimizers import Adam
import matplotlib.pyplot as plt
from keras.callbacks import EarlyStopping
import warnings

warnings.filterwarnings("ignore")

# Load the crop dataset from local CSV file
crop_df = pd.read_csv('output.csv', parse_dates=['Price Date'])
crop_df = crop_df.set_index('Price Date')

# Select commodity
# selected_commodity = input("Enter the commodity name: ")
selected_df = crop_df[crop_df['Commodity'] == 'Bajra(Pearl Millet/Cumbu)']

# Extract the target variable
target = 'Modal Price (Rs./Quintal)'
data = selected_df[target].values.reshape(-1, 1)

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)

# Split the data into training and testing sets
split_index = int(len(data_scaled) * 0.8)
train = data_scaled[:split_index]
test = data_scaled[split_index:]

# Prepare the data for LSTM
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

time_steps = 30  # Number of time steps to look back
X_train, y_train = create_dataset(train, train, time_steps)
X_test, y_test = create_dataset(test, test, time_steps)

# Reshape input to be [samples, time steps, features]
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))

# Early stopping callback
early_stopping = EarlyStopping(patience=10, restore_best_weights=True)

# Define the LSTM model
model = Sequential([
    LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)),
    LSTM(units=50, return_sequences=False),
    Dense(units=1)
])

# Compile the model with a different learning rate
optimizer = Adam(learning_rate=0.0001)
model.compile(optimizer=optimizer, loss='mean_squared_error')

# Train the model with early stopping
history = model.fit(X_train, y_train, epochs=100, batch_size=32, verbose=1, callbacks=[early_stopping])

# Predict on the testing data
predicted_prices_scaled = model.predict(X_test)
predicted_prices = scaler.inverse_transform(predicted_prices_scaled)

# Calculate the RMSE
rmse = np.sqrt(np.mean(np.square(y_test - predicted_prices_scaled)))
print(f'RMSE: {rmse:.2f}')

# Ensure both arrays have the same length for plotting
length_to_use = min(len(selected_df.index[-len(test):]), len(predicted_prices))

# Plotting
plt.figure(figsize=(12, 6))

# Plot actual prices
plt.plot(selected_df.index[-len(test):][:length_to_use], selected_df['Modal Price (Rs./Quintal)'][-len(test):][:length_to_use], label='Actual Prices')

# Plot predicted prices for the test set
plt.plot(selected_df.index[-len(test):][:length_to_use], predicted_prices[:length_to_use], label='Predicted Prices (Test Set)')

# Predict future prices for the next forecast_steps
forecast_steps = int(input("Enter the number of days for future prediction: ")) 
future_dates = pd.date_range(start=selected_df.index[-1], periods=forecast_steps + 1, freq='D')[1:]
last_sequence = X_test[-1]
predicted_prices_future = []

for _ in range(forecast_steps):
    last_sequence_reshaped = last_sequence.reshape((1, time_steps, 1))
    next_price_scaled = model.predict(last_sequence_reshaped)[0][0]
    next_price = scaler.inverse_transform([[next_price_scaled]])[0][0]
    predicted_prices_future.append(next_price)
    last_sequence = np.roll(last_sequence, -1)
    last_sequence[-1] = next_price_scaled

# Plot predicted prices for the next forecast_steps
plt.plot(future_dates, predicted_prices_future, label=f'Predicted Prices ({forecast_steps} Days Forecast)', linestyle='--')

# Print forecasted prices
print("Forecasted Prices for the next", forecast_steps, "days:")
for i in range(forecast_steps):
    print(future_dates[i].strftime('%Y-%m-%d'), "->", predicted_prices_future[i])

plt.title(f'{selected_commodity} Prices Forecasting with LSTM')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Linear Regression

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the crop dataset from local CSV file
crop_df = pd.read_csv('../dataset/cropPrice.csv', parse_dates=['Price Date'])
crop_df = crop_df.set_index('Price Date')

# selected_commodity = input("Enter the District name: ")
selected_df = crop_df[crop_df['District'] == 'Sangli']

# selected_commodity = input("Enter the Market name: ")
selected_df = selected_df[selected_df['Market'] == 'Sangli']

# selected_commodity = input("Enter the Commodity name: ")
selected_df = selected_df[selected_df['Commodity'] == 'Bajra(Pearl Millet/Cumbu)']

# Extract the target variable
target = 'Modal Price (Rs./Quintal)'
data = selected_df[target].values.reshape(-1, 1)

# Normalize the data
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)

# Split the data into training and testing sets
split_index = int(len(data_scaled) * 0.8)
train = data_scaled[:split_index]
test = data_scaled[split_index:]

# Prepare the data for linear regression
def create_dataset(X, y, time_steps=1):
    Xs, ys = [], []
    for i in range(len(X) - time_steps):
        Xs.append(X[i:(i + time_steps)])
        ys.append(y[i + time_steps])
    return np.array(Xs), np.array(ys)

time_steps = 30  # Number of time steps to look back
X_train, y_train = create_dataset(train, train, time_steps)
X_test, y_test = create_dataset(test, test, time_steps)

# Flatten X_train and X_test
X_train = X_train.reshape(X_train.shape[0], -1)
X_test = X_test.reshape(X_test.shape[0], -1)

# Fit linear regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Predict on the testing data
predicted_prices_scaled = model.predict(X_test)
predicted_prices = scaler.inverse_transform(predicted_prices_scaled)

# Calculate the RMSE
rmse = np.sqrt(np.mean(np.square(y_test - predicted_prices_scaled)))
print(f'RMSE: {rmse:.2f}')

# Ensure both arrays have the same length for plotting
length_to_use = min(len(selected_df.index[-len(test):]), len(predicted_prices))

# Plotting
plt.figure(figsize=(12, 6))

# Plot actual prices
plt.plot(selected_df.index[-len(test):][:length_to_use], selected_df['Modal Price (Rs./Quintal)'][-len(test):][:length_to_use], label='Actual Prices')

# Plot predicted prices for the test set
plt.plot(selected_df.index[-len(test):][:length_to_use], predicted_prices[:length_to_use], label='Predicted Prices (Test Set)')

# Predict future prices for the next forecast_steps
forecast_steps = int(input("Enter the number of days for future prediction: ")) 
future_dates = pd.date_range(start=selected_df.index[-1], periods=forecast_steps + 1, freq='D')[1:]
last_sequence = X_test[-1]
predicted_prices_future = []

for _ in range(forecast_steps):
    next_price_scaled = model.predict([last_sequence])[0]
    next_price_scaled = np.reshape(next_price_scaled, (-1, 1))
    next_price = scaler.inverse_transform(next_price_scaled)[0][0]
    predicted_prices_future.append(next_price)
    last_sequence = np.roll(last_sequence, -1)
    last_sequence[-1] = next_price_scaled


# Plot predicted prices for the next forecast_steps
plt.plot(future_dates, predicted_prices_future, label=f'Predicted Prices ({forecast_steps} Days Forecast)', linestyle='--')

# Print forecasted prices
print("Forecasted Prices for the next", forecast_steps, "days:")
for i in range(forecast_steps):
    print(future_dates[i].strftime('%Y-%m-%d'), "->", predicted_prices_future[i])

plt.title(f'Prices Forecasting with Linear Regression')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
